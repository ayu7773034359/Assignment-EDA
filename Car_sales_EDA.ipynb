2.Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing
trends?

```python
df = df.dropna(subset=['Company', 'Price ($)'])
df['Price($)'] = pd.to_numeric(df['Price ($)'], errors='coerce')

price_variation = df.groupby('Company')['Price ($)'].std().sort_values(ascending=False)

top_brand = price_variation.idxmax()
top_std = price_variation.max()
print(f"Company with highest price variation: {top_brand} (Std Dev = {top_std:.2f})")

# Optional: Plot the top 10 brands by price variation
price_variation.head(10).plot(kind='barh', figsize=(10,6), color='skyblue')
plt.xlabel("Standard Deviation of Price ($)")
plt.title("Top 10 Car Company with Highest Price Variation")
plt.gca().invert_yaxis()
plt.show()
```

9.How do car prices vary based on the customer’s annual income bracket?

```python
df = df.dropna(subset=['Annual Income', 'Price ($)'])

# Ensure 'Annual Income' and 'Price ($)' are numeric
# Use 'Price ($)' based on the provided column list and previous successful code
df['Annual Income'] = pd.to_numeric(df['Annual Income'], errors='coerce')
df['Price ($)'] = pd.to_numeric(df['Price ($)'], errors='coerce')

# Drop rows where conversion to numeric failed
df = df.dropna(subset=['Annual Income', 'Price ($)'])

# --- Create the 'Income Bracket' column ---
# Define bins and labels for income brackets
# Adjust these based on your data distribution and analysis needs
income_bins = [0, 30000, 60000, 90000, 120000, np.inf]
income_labels = ['0-30k', '30k-60k', '60k-90k', '90k-120k', '120k+']

# Create the 'Income Bracket' column using pd.cut
# include_lowest=True ensures values at the bin lower bound are included
# If any Annual Income is negative or significantly outside the bins, pd.cut might result in NaN,
# which is handled by the subsequent dropna.
df['Income Bracket'] = pd.cut(df['Annual Income'], bins=income_bins, labels=income_labels, right=False, include_lowest=True)
# --- End of creating 'Income Bracket' ---

# Drop rows where 'Income Bracket' is NaN (e.g., from NaN in Annual Income)
# This line is now valid because 'Income Bracket' has been created.
df = df.dropna(subset=['Income Bracket'])

# Group by 'Income Bracket' and calculate the mean 'Price ($)'
# Use reindex to ensure all defined income_labels are present in the output
# Fill missing values with 0 for clarity in printout if a bracket has no data
average_price_by_income = df.groupby('Income Bracket')['Price ($)'].mean().reindex(income_labels).fillna(0)

print("Average Car Price by Customer Annual Income Bracket:")
print(average_price_by_income)

# Plotting
plt.figure(figsize=(10, 6))
# Ensure the plotting data uses the reindexed series to maintain order
sns.barplot(x=average_price_by_income.index, y=average_price_by_income.values, palette='viridis')
plt.title("Average Car Price by Customer Annual Income Bracket")
plt.xlabel("Annual Income Bracket")
plt.ylabel("Average Price ($)")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

12.Is there any seasonal trend in car sales based on the date of sale

```python
df_filtered = df.dropna(subset=['Date'])

# Ensure 'Date' column is in datetime format
# Use errors='coerce' to turn unparseable dates into NaT (Not a Time)
df_filtered['Date'] = pd.to_datetime(df_filtered['Date'], errors='coerce')

# Drop rows where date conversion failed
df_filtered = df_filtered.dropna(subset=['Date'])

# Extract Year and Month
df_filtered['Sale_Year'] = df_filtered['Date'].dt.year
df_filtered['Sale_Month'] = df_filtered['Date'].dt.month

# Group by Year and Month and count the number of sales
# Use .size() or .count() on any column (like 'Car_id')
monthly_sales = df_filtered.groupby(['Sale_Year', 'Sale_Month']).size().reset_index(name='Sales_Count')

# Sort by Year and Month to plot chronologically
monthly_sales = monthly_sales.sort_values(by=['Sale_Year', 'Sale_Month'])

print("Monthly Sales Counts:")
print(monthly_sales)

# Visualize the monthly sales trend over time
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_sales, x='Sale_Month', y='Sales_Count', hue='Sale_Year', marker='o')

plt.title("Monthly Car Sales Over Time by Year")
plt.xlabel("Month")
plt.ylabel("Number of Sales")
plt.xticks(np.arange(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Year')
plt.tight_layout()
plt.show()


# To specifically look for a recurring seasonal pattern across years,
# you can average sales by month across all years
if df_filtered['Sale_Year'].nunique() > 1: # Only meaningful if data spans multiple years
    average_monthly_sales = df_filtered.groupby('Sale_Month').size().reset_index(name='Average_Sales_Count')
    average_monthly_sales = average_monthly_sales.sort_values(by='Sale_Month')

    print("\nAverage Sales Count by Month (Across All Years):")
    print(average_monthly_sales)

    plt.figure(figsize=(10, 6))
    sns.lineplot(data=average_monthly_sales, x='Sale_Month', y='Average_Sales_Count', marker='o', color='blue')
    plt.title("Average Car Sales by Month (Seasonal Trend)")
    plt.xlabel("Month")
    plt.ylabel("Average Number of Sales")
    plt.xticks(np.arange(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```

14. What is the correlation between car price, engine size, and annual income of customers, and how do these
features interact

```python
df_filtered = df.dropna(subset=['Price ($)', 'Engine', 'Annual Income'])

# Ensure relevant columns are numeric for correlation analysis
df_filtered['Price ($)'] = pd.to_numeric(df_filtered['Price ($)'], errors='coerce')
df_filtered['Annual Income'] = pd.to_numeric(df_filtered['Annual Income'], errors='coerce')

# Check the data type of the 'Engine' column
print(f"Data type of 'Engine' column for correlation analysis: {df_filtered['Engine'].dtypes}")

# If 'Engine' is not numeric, we need to decide how to handle it for correlation.
# Option 1: If it represents displacement (e.g., '1.6', '2.0'), try converting it to numeric.
# Option 2: If it's purely categorical ('V4', 'V6'), correlation matrix won't include it directly.
# Let's attempt to convert 'Engine' to numeric, coercing errors.
df_filtered['Engine_Numeric'] = pd.to_numeric(df_filtered['Engine'], errors='coerce')

# Drop rows where conversion to numeric failed for any of the three key columns
df_filtered = df_filtered.dropna(subset=['Price ($)', 'Annual Income', 'Engine_Numeric'])


# --- Part 1: Correlation ---

# Select the numeric columns for correlation
correlation_columns = df_filtered[['Price ($)', 'Engine_Numeric', 'Annual Income']]

# Calculate the correlation matrix
correlation_matrix = correlation_columns.corr()

print("Correlation Matrix:")
print(correlation_matrix.round(2)) # Round for readability

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Matrix: Car Price, Engine, and Annual Income")
plt.tight_layout()
plt.show()
```

15. How does the average car price vary across different car models and engine types?

```python
df_filtered = df.dropna(subset=['Model', 'Engine', 'Price ($)'])

# Ensure 'Price ($)' is numeric
df_filtered['Price ($)'] = pd.to_numeric(df_filtered['Price ($)'], errors='coerce')

# Drop rows where conversion to numeric failed
df_filtered = df_filtered.dropna(subset=['Price ($)'])

# Group by 'Model' and 'Engine' and calculate the mean 'Price ($)'
average_price_by_model_engine = df_filtered.groupby(['Model', 'Engine'])['Price ($)'].mean()

print("Average Car Price by Model and Engine Type:")
print(average_price_by_model_engine.round(2)) # Round for readability

# To visualize using a heatmap, we need the data in a pivot table format
# Model as rows, Engine as columns, and average price as values
# Handle potential missing combinations (NaNs in the pivot table)
average_price_pivot = average_price_by_model_engine.unstack()

# Check the dimensions of the pivot table
print(f"\nPivot table shape: {average_price_pivot.shape}")

# Visualize the average prices using a heatmap
# Note: Heatmaps can become hard to read with too many unique models or engine types.
if average_price_pivot.shape[0] <= 30 and average_price_pivot.shape[1] <= 30: # Heuristic limit for readability
    plt.figure(figsize=(average_price_pivot.shape[1] * 0.8 + 2, average_price_pivot.shape[0] * 0.5 + 2)) # Adjust figure size dynamically
    sns.heatmap(average_price_pivot, annot=True, fmt=".0f", cmap="viridis", linewidths=.5)
    plt.title("Average Car Price by Model and Engine Type")
    plt.xlabel("Engine Type")
    plt.ylabel("Car Model")
    plt.tight_layout()
    plt.show()
```

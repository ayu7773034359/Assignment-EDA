3.What is the distribution of discounted prices vs. actual prices?

```python
# Ensure the price columns are numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')

# Handle potential NaNs introduced by coercion.
# Filling with 0 might not be the best approach for prices if 0 isn't a valid price.
# A better approach might be to drop rows where prices are NaN or fill with a more
# representative value like the median or mean, but for simplicity here we'll fill with 0.
# Consider your data and decide on the best strategy.
df['discounted_price'].fillna(0, inplace=True)
df['actual_price'].fillna(0, inplace=True)

# Now, visualize the distributions
plt.figure(figsize=(12, 6))

# Histogram for discounted prices
plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
sns.histplot(df['discounted_price'], kde=True, bins=50)
plt.title('Distribution of Discounted Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Histogram for actual prices
plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
sns.histplot(df['actual_price'], kde=True, bins=50, color='orange')
plt.title('Distribution of Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()

# You could also use box plots to see summary statistics and outliers
# plt.figure(figsize=(8, 6))
# sns.boxplot(data=df[['discounted_price', 'actual_price']])
# plt.title('Box Plot of Prices')
# plt.ylabel('Price')
# plt.show()
```

4.How does the average discount percentage vary across categories?

```python
# Ensure price columns are numeric and handle potential NaNs (as done in the previous step)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce').fillna(0)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce').fillna(0)

# Calculate the discount amount
df['discount_amount'] = df['actual_price'] - df['discounted_price']

# Calculate the discount percentage
# Handle cases where actual_price is 0 to avoid division by zero
df['discount_percentage'] = np.where(
    df['actual_price'] > 0,
    (df['discount_amount'] / df['actual_price']) * 100,
    0  # Assign 0% discount if actual price is 0
)

# Group by category and calculate the average discount percentage
average_discount_by_category = df.groupby('category')['discount_percentage'].mean()

# Print the result
print(average_discount_by_category)
```

6.What are the most popular product keywords?

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import re

# Download necessary NLTK data (if you haven't already)
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')
    nltk.download('punkt')


# Ensure the 'product_name' column is a string type
df['product_name'] = df['product_name'].astype(str)

# Combine all product names into a single string
all_product_names = ' '.join(df['product_name'].tolist())

# Tokenize the string into words
words = nltk.word_tokenize(all_product_names)

# Convert to lowercase and remove punctuation
words = [word.lower() for word in words if word.isalpha()] # Keep only alphabetic words

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word not in stop_words]

# Count word frequencies
word_counts = Counter(filtered_words)

# Get the most common words (e.g., top 20)
most_common_keywords = word_counts.most_common(20)

# Print the result
print("Most Popular Product Keywords (based on frequency in product names):")
for word, count in most_common_keywords:
    print(f"{word}: {count}")
```

8.What is the correlation between discounted_price and rating?

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df_cleaned = df.dropna(subset=['discounted_price', 'rating'])
correlation = df_cleaned['discounted_price'].corr(df_cleaned['rating'])
print(f"Correlation between Discounted Price and Rating: {correlation}")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, x='discounted_price', y='rating', alpha=0.5)
plt.title('Scatter Plot of Discounted Price vs. Rating')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.show()

```
